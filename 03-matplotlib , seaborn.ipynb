```python
#                                                 matplotlib
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
%matplotlib inline
```

```python
x=np.linspace(0,5,12)
y=x**2
```

```python
#there are two methods to plot garphs in the matplotlib library
#1st functional method
plt.plot(x,y)
plt.title("TEST")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
#subplots uisng method 1st
plt.subplot(2,2,1)
plt.plot(x,y,"r")
plt.subplot(2,2,2,)
plt.plot(y,x,"g")
```

```python
#2nd method 
#oo method (object oriented method)
fig=plt.figure()   #figure variable created
axis=fig.add_axes([0.2,0.5,0.8,0.9]) #fig variable gets axes pf dim: [l,b,w,h]
axis.plot(x,y)
axis.set_title("TITLE 2")
axis.set_xlabel("X AXIS")
axis.set_ylabel("Y AXIS")
```

```python
#subplots using method 2
fig2=plt.figure()
axis1=fig2.add_axes([0.2,0.5,0.8,0.9])
axis2=fig2.add_axes([0.3,0.95,0.3,0.3])
axis1.plot(x,y,"r")
axis1.set_title("OUTER GRAPH")
axis2.plot(y,x,"b")
axis1.set_title("INNER GRAPH")
```

```python
#MOST EASY TO USE IS FUNCTIONAL METHOD but good method is OO METHOD as it gives flexibility
#2nd way to make subplots using oo method is 
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("1st garph")
axes[1].plot(y,x)
axes[1].set_title("2nd garph")
plt.tight_layout()   #prevents overlapping of graphs
```

```python
#how to alter figure size and DPI(dots per inch) no. for more detailing
fig3, axes2=plt.subplots(figsize=(12,2))
axes2.plot(x,y)
```

```python
#using legends in the garphs to know which line is which eqution 
fig,axes=plt.subplots(nrows=1,ncols=1,dpi=100)
axes.plot(x,x**2,label="square of no.")
axes.set_title("garph to show how legends work")
axes.plot(x,x**3,label="cube of the number")
plt.legend()      #loc can be altered to chnage the location of the legend added 
plt.tight_layout()
```

```python
import seaborn as sns
%matplotlib inline 
```

```python
tips=sns.load_dataset("tips")               #pre existing dataset from seaborn
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"])     #gives distribution  of data present in column
```

```python
#now to remove kde(kernal density estimation)use:
sns.distplot(tips["total_bill"],kde=False,bins=40)  #bins used to know more details
```

```python
#joint plot to join two distribution and know relation b/w them
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")  #kind can also be changed
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
#pairplot
sns.pairplot(tips, hue="sex")   #very important for instant realiztaion of the data we are adding
#hue takes categorical data and represents different cattegory with differnet colour
```

```python
#categorical plot
# 1st barplot
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
#2nd countplot
sns.countplot(x="sex",data=tips)  #gives count of number of male and females in the data
```

```python
#3rd boxplot
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
# 4th violinplot
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
#5th stripplot
sns.stripplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
#6th swarmplot and both swarmplot and violinplot above it
sns.violinplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
#7th factorplot
#strip can be changed to make all the differnet plots for categorical data discussed above
sns.factorplot(x="day",y="total_bill",data=tips,hue="smoker",kind="strip")
```

```python
#                                       matrix plots
import seaborn as sns
%matplotlib inline 
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc=tips.corr() #correlation of the table or conversion of datset into matrix
```

```python
#heat map
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
#converting flight data into matrix using pivot_table
flight_mat=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
sns.heatmap(flight_mat,cmap="coolwarm")
```

```python
sns.heatmap(flight_mat)
```

```python
sns.heatmap(flight_mat,linecolor="white",linewidths=0.5)
```

```python
sns.clustermap(flight_mat)  #creates the heatmap showing similar row corresponding to each other
```

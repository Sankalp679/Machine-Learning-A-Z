```python
import pandas as pd
import numpy as np
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
%matplotlib inline
```

```python
#initial steps 
init_notebook_mode(connected=True)  #connects to the java notebook
cf.go_offline()      #makes cufflink work offline
```

```python
#creating dataset  for our use
df1=pd.DataFrame(np.random.randn(100,4),columns=["a","b","c","d"])
```

```python
df1.head()
```

```python
df1.plot()   #normal matplotlib plotting
```

```python
df1.iplot()        #interactive plot 
```

```python
df1.iplot(kind="scatter",x="a",y="b")       #plot without specifying mode
```

```python
df1.iplot(kind="scatter",x="a",y="b",mode="markers")
```

```python
#bar plot
df1.iplot(kind="bar")
```

```python
#we can also do
df1.sum().iplot(kind="bar")
```

```python
#making 3d interactive plots
df4=pd.DataFrame({"x":[1,2,3,4,5,6],"y":[20,30,50,70,10,80],"z":[6,5,4,3,2,1]})
```

```python
df4
```

```python
df4.iplot(kind="surface")# intercative surface plot
```

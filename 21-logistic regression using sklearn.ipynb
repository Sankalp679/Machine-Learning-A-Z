```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
x,y=make_classification(n_samples=1000,n_classes=2,n_features=2,n_informative=2,n_redundant=0,n_repeated=0)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
data = np.zeros((x.shape[0],x.shape[1]+1))
data[:,1:] = x
data[:,0] += 1.0
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.2,random_state=42)
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
pred=logreg.predict(x_test)
```

```python
print("accuracy_score:",accuracy_score(pred,y_test)*100,"%")
```

```python
print("confusion_matrix:\n",confusion_matrix(pred,y_test))
```
